```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt 
```

```python
df_csv=pd.read_csv('dailyActivity_merged.csv')
df_csv
```

```python
cols=['Id','ActivityDate','TotalSteps','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes','Calories']
new_df=df_csv[cols]
```

```python
new_df
```

```python
new_df['TotalMinutes'] = new_df['VeryActiveMinutes'] + \
                         new_df['FairlyActiveMinutes'] + \
                         new_df['LightlyActiveMinutes'] + \
                         new_df['SedentaryMinutes']
```

```python
new_df
```

```python
new_df['Totalhours'] = round(new_df.TotalMinutes/60)
```

```python
new_df
```

```python
new_df.info()
```

```python
new_df['ActivityDate'] = pd.to_datetime(new_df['ActivityDate'])
new_df['DayOfWeek'] = new_df['ActivityDate'].dt.day_name()
```

```python
new_df
```

```python
new_df.info()
```

```python
plt.figure(figsize=((8,5)))
plt.hist(new_df.DayOfWeek,bins=7,color='lightskyblue',width=0.6)
plt.grid(True)
plt.xlabel('Day of week')
plt.ylabel('Freq')
plt.title('Freq by day of week')
plt.show
```

```python
new_df.corr(numeric_only='True')
```

```python
sns.heatmap(new_df.corr(numeric_only=True))
plt.figure(figsize=((3,3)))
plt.show
```

```python
plt.figure(figsize=((5,5)))
plt.scatter(new_df.TotalSteps,new_df.Calories,c=new_df.Calories)
median_steps=7405
median_calories=2134
plt.axhline(median_calories,color='blue',label='median of calories')
plt.axvline(median_steps,color='red',label='median of steps')
plt.xlabel('Steps')
plt.ylabel('Calories')
plt.title('Calories by total steps')
plt.legend()
plt.show
```

```python
plt.figure(figsize=((5,5)))
plt.scatter(new_df.Totalhours,new_df.Calories,c=new_df.Calories)
median_hours=24
median_calories=2134
plt.axhline(median_calories,color='blue',label='median of calories')
plt.axvline(median_hours,color='red',label='median of hours')
plt.xlabel('hours')
plt.ylabel('Calories')
plt.title('Calories by  hours')
plt.legend()
plt.show
```

```python
VeryActiveMinutes=new_df.VeryActiveMinutes.sum()
FairlyActiveMinutes=new_df.FairlyActiveMinutes.sum()
LightlyActiveMinutes=new_df.LightlyActiveMinutes.sum()
SedentaryMinutes=new_df.SedentaryMinutes.sum()
minutes=[VeryActiveMinutes,FairlyActiveMinutes,LightlyActiveMinutes,SedentaryMinutes]
labels=['VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','SedentaryMinutes']
plt.figure(figsize=((8,7)))

plt.pie(minutes,labels=labels,autopct='%1.1f',explode=[0,0,0,0.1])
plt.show
```
